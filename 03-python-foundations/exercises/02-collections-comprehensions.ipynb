## 1 · Lists for chat traces

*Explanation:* Chat history is a **list** of dicts — order matters.


````python
messages = [
    {"role": "system", "content": "Be concise."},
    {"role": "user", "content": "Hello"},
]
messages.append({"role": "assistant", "content": "Hi — how can I help?"})
print(messages[-1])
````

## 2 · Dicts for configs

*Explanation:* Merge defaults with overrides via `|` (Python 3.9+).


````python
defaults = {"temperature": 0.7, "top_k": 5}
user_prefs = {"temperature": 0.2}
cfg = defaults | user_prefs
print(cfg)
````

## 3 · Comprehensions

*Explanation:* Compact loops for normalization — prefer readable over clever.


````python
messy = [" Rag ", "", "chunk\n", "docs "]
clean = [s.strip().lower() for s in messy if s.strip()]
print(clean)
````

## Progressive drills — **easy → harder**

Collections power **retrieval pipelines**: pairing roles, drilling into nested tool payloads, filtering IDs cheaply.


### A · Easiest — unpack `(role, text)` tuples

SDKs sometimes hand you pairs—unpack instead of indexing `[0]`/`[1]`.


````python
pair = ("system", "Be factual.")
role, directive = pair
print(role, directive[:2])
````

### B · Medium — safe drill into nested tool JSON

`.get` avoids `KeyError` when model JSON is noisy.


````python
payload = {"tool": {"name": "search", "args": {"q": "hybrid retrieval"}}}
query = payload.get("tool", {}).get("args", {}).get("q", "")
print("query:", query)
````

### C · Harder — comprehension filters chunk IDs

Keep only IDs matching a prefix—cheap gate before hitting the vector DB.


````python
ids = ["chunk_finance_01", "sidebar_ad", "chunk_finance_02"]
finance_chunks = [cid for cid in ids if cid.startswith("chunk_finance_")]
print(finance_chunks)
````
